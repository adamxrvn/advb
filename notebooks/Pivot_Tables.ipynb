Calculation of indicators to be investigated in the dataset

```python
df_clean = df_clean.dropna(subset = ['Маржа_v'])
df_delivered = df_delivered.dropna(subset = ['Маржа_v'])
df_phones = df_phones.dropna(subset = ['Маржа_v'])

```

```python
def generate_table(df, column_name = "МесяцДатыЗаказа"):
    columns_list = [column_name, 'Revenue', 'Absolute Margin', 'Margin', 'Average Margin', 'Average Check',
                                  '# of check', '# of goods',
                                 'Average # of goods in check',
                                 '# of unique clients', 'Value of a Client']

    unique_values = df[column_name].unique()
    total = pd.DataFrame(columns=columns_list)

    for val in unique_values:

        list_of_values = [val]
        filtered_df = df[df[column_name] == val]

        # Revenue
        revenue = int(round(filtered_df['СуммаСтроки_v'].sum()))
        list_of_values.append(revenue)

        # Absolute margin
        absmargin = int(round(filtered_df['Маржа_v'].sum()))
        list_of_values.append(absmargin)

        # Margin
        margin = absmargin/revenue
        list_of_values.append(margin)

        # Average Margin деление
        df_numb_check = filtered_df['СуммаСтроки_v'].unique()
        numb_check = df_numb_check.size
        absmargin = int(round(filtered_df['Маржа_v'].sum()))
        list_of_values.append(absmargin/numb_check)

        # Average check
        avarage_check = int(round(
            filtered_df.groupby(['НомерЗаказаНаСайте']).mean()['СуммаСтроки_v'].mean()))
        list_of_values.append(avarage_check)

        # Number of checks
        list_of_values.append(numb_check)

        # Number of goods
        numb_good = filtered_df['Количество'].sum()
        list_of_values.append(numb_good)

        # Average # of goods in check
        avarage_numb_good = int(round(
            filtered_df.groupby(['НомерЗаказаНаСайте']).count()['СуммаСтроки_v'].mean()))
        list_of_values.append(avarage_numb_good)

        # Number of unique clients
        num_of_customers = filtered_df['Телефон_new'].unique().size
        list_of_values.append(num_of_customers)

        # Client's value
        value_of_customer = int(round(absmargin / num_of_customers))
        list_of_values.append(value_of_customer)

        tmp_df = pd.DataFrame([list_of_values], columns=columns_list)
        total = total.append(tmp_df, ignore_index=True)

    return total
```

Function for counting percentage of bought goods compared to ordered.

```python
def pivot_per(pivot_table_month_del, pivot_table_month_all, a = pivot_table_month_del.columns[0]):
    pivot_n_per = pivot_table_month_del[[a, 'Revenue', 'Absolute Margin', '# of check', '# of goods', '# of unique clients']]
    pivot_n_per['Revenue'] = pivot_n_per['Revenue']/pivot_table_month_all['Revenue']*100
    pivot_n_per['Absolute Margin'] = pivot_n_per['Absolute Margin']/pivot_table_month_all['Absolute Margin']*100
    pivot_n_per['Num of check'] = pivot_n_per['# of check']/pivot_table_month_all['# of check']*100
    pivot_n_per['Num of goods'] = pivot_n_per['# of goods']/pivot_table_month_all['# of goods']*100
    pivot_n_per['Num of unique clients'] = pivot_n_per['# of unique clients']/pivot_table_month_all['# of unique clients']*100

    pivot_n_per.rename(columns={'Revenue':'%Revenue', 'Absolute Margin':'%Absolute Margin', '# of check':'%Num of check',
                       '# of goods':'%Num of goods', '# of unique clients':'%Num of unique clients'}, inplace = True )

    return pivot_n_per
```

```python
pivot_month_per = pivot_per(pivot_table_month_del, pivot_table_month_all, a = pivot_table_month_del.columns[0])
pivot_group2_per = pivot_per(pivot_table_group2_del, pivot_table_group2_all, a = pivot_table_group2_del.columns[0])
pivot_weekday_per = pivot_per(pivot_table_weekday_del, pivot_table_weekday_all, a = pivot_table_weekday_del.columns[0])
pivot_type_per = pivot_per(pivot_table_type_del, pivot_table_type_all, a = pivot_table_type_del.columns[0])
pivot_regions_per = pivot_per(pivot_table_regions_del, pivot_table_regions_all, a = pivot_table_regions_del.columns[0]) 
```
